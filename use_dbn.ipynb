# Tensorflow DBN 

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from collections import Counter
import csv 
from tensorflow.contrib import learn
from sklearn.model_selection import train_test_split
import math
import time
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
opcode_count = pd.read_csv(os.path.join("Opcode_1gram_count.csv"), sep='\t', index_col=0)

data = opcode_count.drop(opcode_count.columns[-1], axis=1)

label = opcode_count['label']

data_list = list(data)
label_list = list(label)

np_data = np.array(data)
np_label = np.array(label)
np_data_float = np_data.astype(np.float32)
np_label_float = np_label.astype(np.float32)

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

label_reshape = np_label_float.reshape(np_label_float.shape[0],1)
one_label = ohe.fit_transform(label_reshape).toarray()
```

```python
opcode_count.head()
```

```python
a=0
repeat_list=[]
for i in range(30):
    a = a+1000
    repeat_list.append(a)
```

```python
X_train_list = []
X_val_list = []
X_test_list = []
y_train_list = []
y_val_list = []
y_test_list = []
for i in repeat_list:
    X_train, X_test, y_train, y_test = train_test_split(np_data_float, one_label, test_size=0.2, random_state=i)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=i)
    X_train_list.append(X_train)
    X_val_list.append(X_val)
    X_test_list.append(X_test)
    y_train_list.append(y_train)
    y_val_list.append(y_val)
    y_test_list.append(y_test)
```

```python
tf.set_random_seed(1000)
np.random.seed(1000)
```

```python
import RBM
```

```python
import sigmoid_DBN
```

```python
sigmoid_xavier_train=[]
sigmoid_xavier_val=[]
sigmoid_xavier_time_l=[]
sigmoid_xavier_loss=[]
sigmoid_xavier_pred=[]
sigmoid_xavier_all_time_l=[]
sigmoid_xavier_pre_l=[]
sigmoid_xavier_recall_l=[]
sigmoid_xavier_f1_l=[]
sigmoid_xavier_auc_l=[]

for i in range(30):
    all_start=time.time()
    tf.reset_default_graph()
    RBM_hidden_sizes = [256, 128] # create 4 layers of RBM with size 1500, 700, 400 and 100

    inpX = X_train_list[i]


    rbm_list = []


    input_size = inpX.shape[1]  # 2304

    for i, size in enumerate(RBM_hidden_sizes):
        print('RBM: ',i,' ',input_size,'->',size)
        rbm_list.append(RBM.RBM(input_size,size))  #2304,1500   1500, 700   700,400
        input_size = size
        
    init = tf.global_variables_initializer()
    pre_w = []
    pre_w1 = []
    pre_time = time.time()
#     inpX = X_train_list[i]
    for rbm in rbm_list:
        print('New RBM:')
    
        with tf.Session() as sess:
            sess.run(init)
            rbm.set_session(sess)
            err = rbm.fit(inpX, 1)         # input, epoch
            inpX_n = rbm.rbm_output(inpX)    
            print(inpX_n.shape)
            inpX = inpX_n
    pre_end_time = time.time() - pre_time
    print("pre_end_time : ",pre_end_time)
    
    cal_time = time.time()
    nNet = sigmoid_DBN.sigmoid_DBN(RBM_hidden_sizes, X_train_list[i], y_train_list[i], epochs = 100)
    with tf.Session() as sess:
    #Initialize Variables
        sess.run(tf.global_variables_initializer())
        nNet.set_session(sess)
        nNet.load_from_rbms(RBM_hidden_sizes, rbm_list)
        train_list, val_list,time_list, loss_list = nNet.train(X_val_list[i], y_val_list[i])
        pred_list,pre,recall,f1, auc = nNet.predict_ex(X_test_list[i],y_test_list[i])
        
        sigmoid_xavier_train.append(train_list)
        sigmoid_xavier_val.append(val_list)
        sigmoid_xavier_time_l.append(time_list)
        sigmoid_xavier_loss.append(loss_list)
        sigmoid_xavier_pred.append(pred_list)
        sigmoid_xavier_pre_l.append(pre)
        sigmoid_xavier_recall_l.append(recall)
        sigmoid_xavier_f1_l.append(f1)
        sigmoid_xavier_auc_l.append(auc)
        
    end_time = time.time() - cal_time
    print("end_time : " ,end_time)
    RBM_hidden_sizes = [None]
    all_end = time.time() - all_start
    sigmoid_xavier_all_time_l.append(all_end)
    
```
